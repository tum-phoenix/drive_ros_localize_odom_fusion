```python
import pymongo
import pprint
import matplotlib.pyplot as plt
```

# connect to db
connect to mongo db and get cursor to best x results

```python
experiment = '2018-01-09-02-05-54'
result_count = 200

client = pymongo.MongoClient('localhost', 1234)
db = client[experiment]
jobs = db['jobs']
cursor = jobs.find({ "result.status": "ok" }, limit=result_count).sort( "result.loss", pymongo.ASCENDING )
```

# save to list
save data in lists

```python
x = list()
y = list()
v = list()
a = list()
omega = list()
theta = list()
loss = list()

for job in cursor:
    x.append(job['misc']['vals']['x'])
    y.append(job['misc']['vals']['y'])
    v.append(job['misc']['vals']['v'])
    a.append(job['misc']['vals']['a'])
    omega.append(job['misc']['vals']['omega'])
    theta.append(job['misc']['vals']['theta'])
    loss.append(job['result']['loss'])
```

# fancy visualization

```python
# x and y
plt.scatter(x, loss)
plt.scatter(y, loss)
plt.xlabel('variances pos_x=blue & pos_y=orange')
plt.ylabel('loss')
plt.show() 
```
